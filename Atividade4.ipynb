```python
import pandas as pd
from tkinter import Tk
from tkinter.filedialog import askopenfilename
from matplotlib import pyplot as plt
```

```python
def calcular(dataf):
    piloto = pd.DataFrame()
    
    piloto = dataf
    
    
    media = piloto[['Obs1','Obs2','Obs3','Obs4','Obs5','Obs6','Obs7']].mean(axis=1)
    desvio = piloto[['Obs1','Obs2','Obs3','Obs4','Obs5','Obs6','Obs7']].std(axis=1)
    colunas = len(dataf.iloc[0].to_list())
    desvio_amostral = desvio.mean()/(colunas**0.5)
    
    
    LM = media.mean()
    LSC = LM + (3*desvio_amostral)
    LIC = LM - (3*desvio_amostral)
    LSC_2sigma = LM + (2*desvio_amostral)
    LIC_2sigma = LM - (2*desvio_amostral)
    
    
    piloto.insert(loc=colunas,column='MÃ©dia da Amostra',value=media)
    piloto.insert(loc=colunas+1,column='LM',value=LM)
    piloto.insert(loc=colunas+2,column='LSC',value=LSC)
    piloto.insert(loc=colunas+3,column='LIC',value=LIC)
    piloto.insert(loc=colunas+4,column='LSC_2sigma',value=LSC_2sigma)
    piloto.insert(loc=colunas+5,column='LIC_2sigma',value=LIC_2sigma)
    
    return piloto
```

```python
Tk().withdraw()
arquivo_teste = askopenfilename()
dados_teste = pd.read_csv(arquivo_teste, delimiter =';',decimal=',')
```

```python
dataf = pd.DataFrame()
dataf = dados_teste
dataf
```

```python
dataf = dataf.dropna()
dataf
```

```python
piloto = pd.DataFrame()
piloto = calcular(dataf)
piloto
```

```python
resposta = 1

tamanho = len(dataf['Obs1'].to_list())
    
amostras = []
while resposta == 1:
    
    media = piloto['MÃ©dia da Amostra']
    LM = piloto['LM'].iloc[0]
    LSC = piloto['LSC'].iloc[0]
    LIC = piloto['LIC'].iloc[0]
    
    t = 0
    resposta=0
    while t <= (tamanho-1) and (t not in amostras):        
        media_amostra = piloto['MÃ©dia da Amostra'].iloc[t]
        
        if media_amostra > LSC or media_amostra < LIC:
            amostras.append(t)
            resposta = 1
           
        t+=1
    if resposta == 1:
        piloto = pd.DataFrame()
        dataf = pd.DataFrame()
        dataf = dados_teste
        dataf = dataf.dropna()
        for x in amostras:
            dataf = dataf.drop(index=x)
        
        tamanho = len(dataf['Obs1'].to_list())       
        piloto = calcular(dataf)        
```

```python
piloto
```

```python
piloto['MÃ©dia da Amostra'].plot(figsize=(10,7),color='blue',marker='s')
piloto['LM'].plot(color='green',linestyle='--')
piloto['LSC'].plot(color='red',linestyle='--')
piloto['LIC'].plot(color='red',linestyle='--')
piloto['LSC_2sigma'].plot(color='yellow',linestyle='--')
piloto['LIC_2sigma'].plot(color='yellow',linestyle='--')
plt.title('Controle das Amostras')
plt.xlabel('Amostras')
plt.ylabel('MÃ©dia')
plt.legend(bbox_to_anchor=(1,1),loc=0)
plt.grid(True)
plt.show()
```

```python
LM = piloto['LM'].iloc[0]
LSC = piloto['LSC'].iloc[0]
LIC = piloto['LIC'].iloc[0]
linhas = len(media_producao.to_list())
amostras_fora = []
```

```python
t=0
while t< linhas:
    media = media_producao.iloc[t]
    if media >LSC or media <LIC:
        
        amostras_fora.append(dados_producao['Amostra'].iloc[t])
    t+=1
qtd= len(amostras_fora)
```

```python
arquivo=open('Resultado_producao.txt','w')
arquivo.write('LSC: '+str(LSC)+'\n')
arquivo.write('LM: '+str(LM)+'\n')
arquivo.write('LIC: '+str(LIC)+'\n')
arquivo.write('Quantidade de Amostras Fora do Limite: ' +str(qtd)+'\n')
arquivo.write('Amostras Fora do Limite:\n')
for t in amostras_fora:    
    arquivo.write(str(t)+'\n')


arquivo.close()
```
